```python
import pandas as pd #Thư viện chính để xử lý và phân tích dữ liệu.
import numpy as np #Thư viện hỗ trợ tính toán số học.
import matplotlib.pyplot as plt #'matplotlib.pyplot' và 'seaborn': Thư viện vẽ biểu đồ.
import statsmodels.api as sm #'scipy.stats' và 'statsmodels.api': Thư viện hỗ trợ phân tích thống kê.
import seaborn as sns
import scipy.stats as stats
#'pd.read_csv("processed_dulieuxettuyendaihoc.csv", index_col = "STT")': Đọc dữ liệu từ tệp CSV có tên 
# "processed_dulieuxettuyendaihoc.csv" và sử dụng cột "STT" làm chỉ số của DataFrame.
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv", index_col = "STT")
#'df': Biến chứa DataFrame được đọc từ tệp CSV.
print(df.head(10))
```

```python
def get_mean(scores): #'def get_mean(scores):': Đây là cách định nghĩa một hàm trong Python. Tên hàm là 'get_mean' và nó nhận một tham số đầu vào là 'scores'.
    return sum(scores)/len(scores)
    #'sum(scores)': Hàm 'sum()' của Python được sử dụng để tính tổng tất cả các phần tử trong danh sách 'scores'.
```

```python
result = df['DH1'].sum()
#'df['DH1'].sum()': Lấy cột DH1 từ DataFrame 'df' và tính tổng tất cả các giá trị trong cột này.
#'sum()': Đây là hàm của pandas để tính tổng các giá trị trong một Series (một cột của DataFrame).
#'result = df['DH1'].sum()': Kết quả tính tổng của cột DH1 được lưu vào biến 'result'.
print("\nThe result of applying the function on the dataframe:\n")
print(result)
```

```python
def Q1(g):
    return np.percentile(g,25)
#'np.percentile(g, 25)': Sử dụng hàm 'percentile' của NumPy để tính giá trị tại vị trí 25% trong mảng 'g'.
```

```python
def Q2(g):
    return np.percentile(g,50)
```

```python
def Q3(g):
    return np.percentile(g,75)
```

```python
def std(g):
    return np.std(g)
```

```python
pd.pivot_table(data=df ,index='KT', values='DH1', aggfunc=['sum','mean','median', 'min', 'max',std,Q1,Q2,Q3])
#'pd.pivot_table()': Hàm tạo bảng tổng hợp (pivot table) từ DataFrame.
#'data=df': Sử dụng DataFrame df.
#'index='KT'': Chỉ định cột KT làm chỉ mục của bảng tổng hợp.
#'values='DH1'': Chỉ định cột DH1 là cột được tính toán.
#'aggfunc=['sum', 'mean', 'median', 'min', 'max', 'std', Q1, Q2, Q3]': Danh sách các hàm tính toán cần thực hiện:
#'sum': Tổng.
#'mean': Giá trị trung bình.
#'median': Trung vị.
#'min': Giá trị nhỏ nhất.
#'max': Giá trị lớn nhất.
#'std': Độ lệch chuẩn.
#'Q1, Q2, Q3': Các hàm tùy chỉnh để tính các tứ phân vị (cần được định nghĩa trước).
```
